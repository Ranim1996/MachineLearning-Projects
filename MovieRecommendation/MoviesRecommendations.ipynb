# Movie Recommendation System

This notebook is created to practice with Finding movie similarties using Cosine Similaritya dn Item-Based Collaborative Filtering.

```python
import pandas as pd
import numpy as np
import operator

ratings_cols = ['user_id', 'movie_id', 'rating']
ratings = pd.read_csv('datasets/u.data', sep='\t', names=ratings_cols, usecols=range(3), encoding="ISO-8859-1")

movies_cols = ['movie_id', 'title']
movies = pd.read_csv('datasets/u.item', sep='|', names=movies_cols, usecols=range(2), encoding="ISO-8859-1")

ratings = pd.merge(movies, ratings)
ratings.head()
```

```python
movieRatings = ratings.pivot_table(index=['user_id'],columns=['title'],values='rating')
movieRatings.head()
# NAN values mean that the user did not rate this movie or watch yet
```

# Cosine Similarity:

```python
starWarsRatings = movieRatings['Star Wars (1977)']
starWarsRatings.head()
# get users rated movie star wars
```

```python
similarMovies = movieRatings.corrwith(starWarsRatings)
similarMovies = similarMovies.dropna()
df = pd.DataFrame(similarMovies)
df.head(10)
# corrwith computes the pairwise correlation of 'Star Wars' vector of user rating with every other movie 
# drop any results that have no data
# create new DataFrame of movies and their correlation score (similarity) to Star Wars:
```

```python
movieStats = ratings.groupby('title').agg({'rating': [np.size, np.mean]})
movieStats.head()
```

```python
popularMovies = movieStats['rating']['size'] >= 100
movieStats[popularMovies].sort_values([('rating', 'mean')], ascending=False)[:15]
# gets only movies rated >= than 100
```

```python
df = movieStats[popularMovies].join(pd.DataFrame(similarMovies, columns=['similarity']))
df.head()
# create new dataframe containg the similarMovies cols
```

```python
df.sort_values(['similarity'], ascending=False)[:10]
```

# Item_Based Collaborative Filtering:

Get movies based on item collaborative filtering.

```python
corrMatrix = movieRatings.corr()
corrMatrix.head()
# get correlation between two movies based on the user rating
```

```python
corrMatrix = movieRatings.corr(method='pearson', min_periods=100)
corrMatrix.head()
# get correlation based on pearson correlation where at least 100 people rated a movie
```

```python
myRatings = movieRatings.loc[0].dropna()
myRatings
# get the user with id 0 with the movies he rated and drop the nan values
```

```python
# fins similar movies to the user with id 0 from the correlation matrix
simCandidates = pd.Series()
for i in range(0, len(myRatings.index)):
    print ("Adding similartiess for " + myRatings.index[i] + "...")
    # Retrieve similar movies to this one that the user rated
    sims = corrMatrix[myRatings.index[i]].dropna()
    # Now scale its similarity by how well the user rated this movie
    sims = sims.map(lambda x: x * myRatings[i])
    # Add the score to the list of similarity candidates
    simCandidates = simCandidates.append(sims)
    
#results:
print ("sorting...")
simCandidates.sort_values(inplace = True, ascending = False)
print (simCandidates.head(10))
```

```python
simCandidates = simCandidates.groupby(simCandidates.index).sum() # remove duplication
simCandidates.sort_values(inplace = True, ascending = False)
simCandidates.head(10)
```

```python
filteredSims = simCandidates.drop(myRatings.index)
filteredSims.head(10)
# drop the movies that user with id 0 rated
```

# KNN (K-Nearest-Neighbors)

Predict the rating of a movie by looking at the 10 movies which are closest to its terms of kinds and popularity.

```python
r_cols = ['user_id', 'movie_id', 'rating']
movieRatings = pd.read_csv('datasets/u.data', sep='\t', names=r_cols, usecols=range(3))
movieRatings.head()
```

```python
movieProperties = ratings.groupby('movie_id').agg({'rating': [np.size, np.mean]})
movieProperties.head()
# group by movie ID, and compute the total number of ratings and the mean rating for every movie
```

```python
movieNumRatings = pd.DataFrame(movieProperties['rating']['size'])
movieNormalizedNumRatings = movieNumRatings.apply(lambda x: (x - np.min(x)) / (np.max(x) - np.min(x)))
movieNormalizedNumRatings.head()
# the raw number of ratings isn't very useful for computing distances between movies
# so create a new DataFrame that contains the normalized number of ratings. 
# a value of 0 means nobody rated it, and a value of 1 or almost 1 will mean it's the most popular movie.
```

```python
movieDict = {}
with open(r'datasets/u.item', encoding="ISO-8859-1") as f:
    temp = ''
    for line in f:
        fields = line.rstrip('\n').split('|')
        movieID = int(fields[0])
        name = fields[1]
        kinds = fields[5:25]
        kinds = map(int, kinds)
        movieDict[movieID] = (name, np.array(list(kinds)), movieNormalizedNumRatings.loc[movieID].get('size'), movieProperties.loc[movieID].rating.get('mean'))
# put everything into one big Python dictionary called movieDict. 
# each entry will contain the movie name, list of genre values (0 or 1), 
# the normalized popularity score, and the average rating for each movie
```

```python
print(movieDict[1]) # print first raw
```

```python
from scipy import spatial

def ComputeDistance(a, b):
    genresA = a[1]
    genresB = b[1]
    genreDistance = spatial.distance.cosine(genresA, genresB)
    popularityA = a[2]
    popularityB = b[2]
    popularityDistance = abs(popularityA - popularityB)
    return genreDistance + popularityDistance
    
ComputeDistance(movieDict[2], movieDict[4])

# compute the "distance" between two movies based on how similar their kinds are, and how similar their popularity is. 
```

```python
print(movieDict[2])
print(movieDict[4])
# the higher distance the less similar check the kinds
```

```python
def getNeighbors(movieID, K):
    distances = []
    for movie in movieDict:
        if (movie != movieID):
            dist = ComputeDistance(movieDict[movieID], movieDict[movie])
            distances.append((movie, dist))
    distances.sort(key=operator.itemgetter(1))
    neighbors = []
    for x in range(K):
        neighbors.append(distances[x][0])
    return neighbors

K = 10
avgRating = 0
neighbors = getNeighbors(1, K)
for neighbor in neighbors:
    avgRating += movieDict[neighbor][3]
    print (movieDict[neighbor][0] + " " + str(movieDict[neighbor][3]))
    
avgRating /= K

# compute the distance between (Toy Story) and all of the movies in the data set. 
# sort those by distance, and print out the K nearest neighbors:
```
